```python
import pandas as pd
```

```python
final_planet_df = pd.read_csv('/content/PRO-NASA-Exoplanet-Scraped-Data/final_scraped_data.csv')

archive_planet_df = pd.read_csv('/content/PRO-NASA-Exoplanet-Scraped-Data/PSCompPars.csv')
```

```python
final_planet_df.head()
```

```python
final_planet_df.drop(columns=['Unnamed: 0'], inplace=True)
final_planet_df.head()
```

```python
final_planet_df.shape
```

```python
archive_planet_df["pl_name"] = archive_planet_df["pl_name"].str.lower()
archive_planet_df.head()
```

```python
archive_planet_df = archive_planet_df.sort_values("pl_name")
archive_planet_df.tail(10)
```

```python
merge_planets_df = pd.merge(final_planet_df,archive_planet_df,on = "id")
merge_planets_df.shape
```
