```python
import numpy as np
import random

rewards = [
    [-1,-1,-1,-1,0,-1],
    [-1,-1,-1,0,-1,100],
    [-1,-1,-1,0,-1,-1],
    [-1,0,0,-1,-1,-1],
    [0,-1,-1,-1,-1,100],
    [-1,-1,-1,-1,-1,100]
    ]

def set_initial_state():
  return np.random.randint(0,6)

def get_action(current_state,reward_matrix):
  available_action=[]
  #print(f"reward matrix: \n {reward_matrix}")
  for action in enumerate(reward_matrix[current_state]):
    if action[1] != -1:
      available_action.append(action[0])
  #for i in range(len(reward_matrix[current_state])):
    #if reward_matrix[current_state][i] != -1:
      #available_action.append(i)
  choose_action = random.choice(available_action)
  print(f"random choice of action from {available_action} is {choose_action}")
  return choose_action

initial_state = set_initial_state()
print(f"initial_state: {initial_state}")

while initial_state != 5:
  action = get_action(initial_state,rewards)
  initial_state = action
  print(f"initial_state: {initial_state}")

print("You have done it!!!!")
```
